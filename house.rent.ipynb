```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data={
    'bedrooms':[2,34,5,6,7,8,9],
    'size':[1500,4533,5000,8000,4999,5000,6000],
    'price':[2345,4553,7535,6553,3576,6666,7753]

}
df=pd.DataFrame(data)
df.dtypes
```

```python
features=df[["bedrooms","size"]]
features
label=df["price"]
label
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
prid_value=lr.predict(x_test)
prid_value
```

```python
x=int(input("enter number of bedroom"))
y=int(input("enter size of bedroom"))
print(f"price of home is{lr.predict([[x,y]])}")
```
